```python
time_zones = [rec['tz'] for rec in records] # list conprehension
# 오류가 뜨는 이유는 모든 리스트가 'tz' 칼럼을 포함하고 있지 않아서 그렇다.
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

```python
# 카운트를 해주는 메소드
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
# 파이썬 표준 라이브러리 사용
from collections import defaultdict
def get_counts2(sequence):
    counts = defaultdict(int) # values will initialize to 0
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
# 두 개의 카운트 함수 비교
counts1 = get_counts(time_zones)
counts2 = get_counts2(time_zones)

print ('counts1 : ', counts1)
print ('counts2 : ', counts2)
```

```python
# 가장 많이 등장하는 상위 10개의 표준시간대
# 딕셔너리의 요소 "for" 한개의 딕셔너리 "in" 딕셔너리들
def top_counts(count_dict, n=10):
    value_key = [(count, tz) for tz, count in count_dict.items()]
    # 튜플 형태로 묶어서 계산, 튜플로 묶이면서 위치를 바꿔준 모습
    # .items() 메소드는 키값을 추출해주는 듯
    value_key.sort()
    return value_key[-n:] # n이 10이므로 맨 끝에서 10개만 select
```

```python
print (top_counts(counts1))
```

```python
# 상위 10개 추출의 파이썬 표준 라이브러리 활용 사례
from collections import Counter
counts = Counter(time_zones)
counts.most_common(10) # 직접 정의하는것보다 내림차순으로 튜플로 정리된 모습
```

```python
### agent 문자열이 'Windows'면 윈도우 사용자, 아니면 제외 ###
cframe = frame[frame.a.notnull()]

oper_sys = np.where(cframe['a'].str.contains('Windows'),
                    'Windows', 'Not Windows')
# where : 찾을 문자열, True 일경우 문자열, False 일경우 문자열

Series(oper_sys[:5])
```

```python
# 표준시간대를 기준으로 oper_sys를 groupping 한다.
by_tz_os = cframe.groupby(['tz', oper_sys])
```

```python
agg_counts1 = by_tz_os.size().unstack().fillna(0)
# value_counts 처럼 그룹별 합계는 size로 구할 수 있다.
# unstack 함수로 표로 재배치한다.
print (agg_counts1[:10])

agg_counts2 = by_tz_os.size().fillna(0)

print (agg_counts2[:10])

'''
unstack을 한 것과 안 한 것을 비교해보니 unstack의 의미를 제대로 알 수 있을 것 같다.
stack을 하게 되면 아래로 쭉 내려가는 표가 되고,
unstack을 하게 되면 칼럼별로 정리가 된다. 말그대로 stack을 풀고 정리가 되는 것 같음
'''
```

```python
agg_counts1
```

```python
### 전체 표준시간대의 순위를 모아보고 싶을 때 ###
indexer = agg_counts1.sum(1).argsort()
# argsort() 오름차순으로 정렬
indexer[:10]
```

```python
count_subset = agg_counts1.take(indexer)[-10:]

count_subset

'''
굉장히 중요한 점,
argsort()를 했음에도 불구하고 ABC로 정렬이 되었을 뿐, 순위가 정렬되지는 않았다.
이는, 논리적으로 정렬이 된 것일 뿐 row의 순서에는 변함이 없다는 뜻.
take를 사용하게 되면 논리적으로 정렬된 순서를 그대로 선택하고 [n:]에 원하는 만큼
가져오게 된다.
'''
```

```python
count_subset.plot(kind='barh', stacked=True)
```

```python
### 위의 그래프는 윈도우 사용자의 상대비율을 알기 힘들기 때문에,
### 각 행에서 총합을 1로 정규화 한 후 표를 만들면 쉽게 확인할 수 있다. ###
normed = count_subset.div(count_subset.sum(1), axis=0)

normed.plot(kind='barh', stacked=True)
```
